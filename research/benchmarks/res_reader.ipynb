```python
import pickle
import matplotlib
import numpy as np
from matplotlib import pyplot as plt
```

```python
BATCHES = 214
EPOCH = 500
DUMP_FOR = 10
REPS = 5
```

# ResNet

```python
for rep in range(REPS):
    loss = []
    for it in range(DUMP_FOR, EPOCH+1, DUMP_FOR):
        try:
            with open('ClassicResNetResearch_500ep/results/layout_cnacna-model_ResNet18/{}/train_{}'.format(rep, BATCHES*it), 'rb') as f:
                res = pickle.load(f)
        except FileNotFoundError:
            it = it - DUMP_FOR
            break
        loss.extend(res['loss'])
    loss = np.split(np.array(loss), it)
    loss = np.mean(loss, axis=1)
    plt.plot(loss[200:])
```

```python
for rep in range(REPS):
    loss = []
    for it in range(DUMP_FOR, EPOCH+1, DUMP_FOR):
        try:
            with open('CustomResNetResearch_500ep/results/blocks_[2, 2, 2, 2, 2, 2, 2]-filters_[4, 4, 8, 8, 16, 16, 20]-input_filters_8-layout_cna-model_ResNetCustom/{}/f_score_{}'.format(rep, BATCHES*it), 'rb') as f:
            #with open('ClassicResNetResearch_500ep/results/layout_cnacna-model_ResNet34/{}/f_score_{}'.format(rep, BATCHES*it), 'rb') as f:
                res = pickle.load(f)
                loss.append(res['f_score'])
        except FileNotFoundError:
            loss = loss[:-1]
            break
    plt.plot(loss)
```

```python
with open('CustomResNetResearch_500ep/results/blocks_[1, 1, 1, 1, 1, 1, 1]-filters_[4, 4, 8, 8, 16, 16, 20]-input_filters_8-layout_cna-model_ResNetCustom/0/trainable_variables_2', 'rb') as f:
    res = pickle.load(f)
res['trainable_variables']
```
